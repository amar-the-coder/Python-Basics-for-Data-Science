## now we have loc and iloc

## Accessing data from a Series


Because a Series has two different kinds of indexes in the backend. Just using simple [] doesn't
work anymore. It causes confusion.

So to get rid of this confusion, pandas introducted two attributes.

1. `.loc[]` - use to access data from user defined index
            - while slicing, .loc is inclusive

2. `.iloc[]` - use to access data from the default index
            - while slicing, last value will be excluded.


```python
s2
```

```python
# using loc now -- it always works with user defined functions:-
s2.loc['Jan':'Feb'] 
```

```python
s2.loc['Jan':'March']
```

```python
s2.loc['Feb':]
```

```python
s2
```

```python
# using iloc
s2.iloc[0] # it works on default index
```

```python
s2.iloc[0:3]
```

4 ways that pandas provides of accessing columns.

1. `[]` - this syntax takes values in only a single dimension
2. `.loc[:, column_name]`  - works in two dimension
3. `.iloc[:, coloumn_index]`
4. `file_name[column_name]`
